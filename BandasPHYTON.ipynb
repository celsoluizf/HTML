```python
import matplotlib.pyplot as plt

banda = ['Guns_N_Roses','Bon_Jovi','Aerosmith','Scorpions','Bryan_Adams']
pessoas = [80, 50, 70, 120, 40]
colours = ['darkred','darkblue','gold','gray','purple']

explode = [0, 0, 0, 0.2, 0]

plt.figure(figsize=(5,5))
plt.pie(pessoas,
        labels = None,
        colors = colours,
        autopct = '%1.0f%%',
        startangle = 90,
        explode = explode,
        shadow = True,
        textprops = {'fontweight':'bold'}
        )

plt.title('Melhores músicos/bandas de todos os tempos', fontweight='bold')

plt.legend(banda,
           title = "Bandas/Músicos",
           loc = "center left",
           bbox_to_anchor=(1, 0.5),
           title_fontsize = '12',
           frameon = True,
           shadow = True,
           fancybox = True,
           facecolor = 'white',
           edgecolor = 'gray',
           labelspacing = 1,
           handletextpad = 1.5,
           borderpad = 1.6
           )

plt.axis('equal')

plt.show()
```
